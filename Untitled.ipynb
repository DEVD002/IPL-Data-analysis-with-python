### importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Getting Data

```python
import pandas_datareader 
import datetime
```

```python
import pandas_datareader.data as web
```

```python
start = datetime.datetime(2012,1,1)
end = datetime.datetime(2021,1,1)
```

```python
tesla = web.DataReader('TSLA','yahoo',start,end)
```

```python
tesla.head()
```

```python
 ford = web.DataReader('F','yahoo',start,end)
```

```python
gm = web.DataReader('GM','yahoo',start,end)
```

```python
ford.head()
```

```python
gm.head()
```

### visualizing data

```python
tesla['Open'].plot(label = 'Tesla', figsize = (12,8), title = 'Opening price')
ford['Open'].plot(label = 'Ford')
gm['Open'].plot(label = 'General Motors')
plt.legend();
```

```python
tesla['Volume'].plot(label = 'Tesla', figsize = (18,10), title = 'Volume Traded')
ford['Volume'].plot(label = 'Ford')
gm['Volume'].plot(label = 'General Motors')
plt.legend();
```

### look like tesla had really big spike in early 2020s.

```python
tesla['Open'].plot(figsize = (20,6))
```

### creating new column for total traded money for each dataframe

```python
tesla['Total_traded'] = tesla['Open']*tesla['Volume']
```

```python
ford['Total_traded'] = ford['Open']*ford['Volume']
gm['Total_traded'] = gm['Open']*gm['Open']
```

```python
tesla['Total_traded'].plot(label = 'Tesla', title= 'Total Traded Money', figsize = (12,8))
ford['Total_traded'].plot(label= 'Ford')
gm['Total_traded'].plot(label = 'General Ford')
plt.legend();
```
